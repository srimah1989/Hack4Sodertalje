```python
# Import modules
import os
import pandas as pd
import folium
import geopandas as gpd
import matplotlib.pyplot as plt
from folium.plugins import MarkerCluster
%matplotlib inline
```

### Set folder PATHS

```python
PATH_notebooks = os.getcwd()
PATH_data = os.path.join(PATH_notebooks[:-10],'data') 
print(PATH_data)
```

### Display list of iles in /data/ folder

```python
os.listdir(PATH_data)
```

### Load Parks Data

```python
df_parks = gpd.read_file(os.path.join(PATH_data, 'green-spaces-geojson-data/parks-sodertalje.json'))
df_parks
```

### Load Natural Reserves data

```python
df_nres = gpd.read_file(os.path.join(PATH_data, 'green-spaces-geojson-data/Natural-reserves-gejson.json'))
df_nres
```

### Load public toilet data 

```python
df_pt = gpd.read_file(os.path.join(PATH_data, 'public-toilets-geojson.json'))
df_pt
```

### Load base map

```python
map_base = folium.Map(location=[59.1978, 17.6202], zoom_start=13)
map_base
```

```python
map_cluster = MarkerCluster().add_to(map_base)
```

```python
for i in range(df_parks.shape[0]):

    Name = df_parks['Namn'][i]
    Lng = df_parks['geometry'][i].x
    Lat = df_parks['geometry'][i].y

    folium.Marker(
        location=[Lat, Lng],
        popup=Name,
        icon=folium.Icon(color="green"),
    ).add_to(map_base)

# Save map
#map_cluster.save("./map_parks_cluster.html")
```

```python
for i in range(df_pt.shape[0]):

    Name = df_pt['Plats'][i]
    Lng = df_pt['geometry'][i].x
    Lat = df_pt['geometry'][i].y

    folium.Marker(
        location=[Lat, Lng],
        popup=Name,
        icon=folium.Icon(icon="cloud"),
    ).add_to(map_base)


# Save map
#map_cluster.save("./map_parks_cluster.html")
```

```python
for i in range(df_nres.shape[0]):

    Name = df_nres['Namn'][i]
    Lng = df_nres['geometry'][i].x
    Lat = df_nres['geometry'][i].y

    folium.Marker(
        location=[Lat, Lng],
        popup=Name,
        icon=folium.Icon(color="red"),
    ).add_to(map_base)

map_base
```
